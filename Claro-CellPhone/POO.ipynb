```python
from colorama import Fore, Style

class Personaje:

    def __init__(self, nombre, fuerza, inteligencia, defensa, vida):
        self.nombre = nombre 
        self.fuerza = fuerza
        self.inteligencia = inteligencia
        self.defensa = defensa
        self.vida = vida
    
    def __str__(self):
        return f""" 
                \n
                Nombre: {Fore.GREEN + Style.BRIGHT} {self.nombre} {Style.RESET_ALL}
                Fuerza: {Fore.GREEN + Style.BRIGHT} {self.fuerza} {Style.RESET_ALL}
                Inteligencia: {Fore.GREEN + Style.BRIGHT} {self.inteligencia} {Style.RESET_ALL}
                Defensa: {Fore.GREEN + Style.BRIGHT} {self.defensa} {Style.RESET_ALL}
                Vida: {Fore.GREEN + Style.BRIGHT} {self.vida} {Style.RESET_ALL}
                """
    
    def subir_vida(self, fuerza=1, inteligencia=1, defensa=1):
        self.defensa += defensa
        self.fuerza += fuerza
        self.inteligencia += inteligencia

    def esta_vivo(self):
        return self.vida > 0 
    
    def morir(self):
        self.vida = 0
        return f'{self.nombre} {Fore.RED + Style.BRIGHT} ¡Muerto! {Style.RESET_ALL}' 
    
    def daño(self, enemigo):
        return self.fuerza - enemigo.defensa
    
    def atacar(self, enemigo):
        daño = self.daño(enemigo)
        enemigo.vida -= daño
        
        print(f'{self.nombre} ataca a {enemigo.nombre} y le hace {daño} puntos de daño')

        if enemigo.esta_vivo():
            print(f'La vida de {enemigo.nombre} es de {enemigo.vida}')
        else:
            enemigo.morir()

class Guerrero(Personaje):
        
    def __init__(self, nombre, fuerza, inteligencia, defensa, vida, arma_maldita):
        super().__init__(nombre, fuerza, inteligencia, defensa, vida)
        self.arma_maldita = arma_maldita
        
    def cambiar_arma_maldita(self):
        opcion = int(input('Elige un Arma maldita: (1) Espada 8 de daño. (2) Chacos 6 de daño. '))
        if opcion == 1:
            self.arma_maldita = 8
        elif opcion == 2:
            self.arma_maldita = 6
        else:
            print("Opción incorrecta")
        
    def __str__(self):
        return super().__str__() + f'Arma_maldita: {Fore.GREEN + Style.BRIGHT} {self.arma_maldita} {Style.RESET_ALL}'

    def daño(self, enemigo):
        return self.fuerza * self.arma_maldita - enemigo.defensa

class Mago(Personaje):

    def __init__(self, nombre, fuerza, inteligencia, defensa, vida, magia_maldita):
        super().__init__(nombre, fuerza, inteligencia, defensa, vida)
        self.magia_maldita = magia_maldita

    def __str__(self):
        return super().__str__() + f'Magia Maldita: {Fore.GREEN + Style.BRIGHT} {self.magia_maldita} {Style.RESET_ALL}'

    def daño(self, enemigo):
        return self.fuerza * self.magia_maldita - enemigo.defensa
```

```python
# Crear objetos de cada Clase
nanami = Personaje(nombre='Nanami', fuerza=16, inteligencia=7, defensa=5, vida=100)
itadori = Guerrero(nombre='Itadori', fuerza=10, inteligencia=3, defensa=10, vida=100, arma_maldita=8)
gojo = Mago(nombre='Gojo', fuerza=10, inteligencia=7, defensa=5, vida=100, magia_maldita=10)

# Polimorfismo
def combate(personaje1, personaje2):
    turno = 0
    while personaje1.esta_vivo() and personaje2.esta_vivo():
        print('\n Turno:', turno)
        print(f'>>>> Acción de {personaje1.nombre}:')
        personaje1.atacar(personaje2)
        if personaje2.esta_vivo():
            print(f'>>>> Acción de {personaje2.nombre}:')
            personaje2.atacar(personaje1)
        turno += 1

    if personaje1.esta_vivo():
        print(f"\n Ha ganado {personaje1.nombre}")
    elif personaje2.esta_vivo():
        print(f"\n Ha ganado {personaje2.nombre}")
    else:
        print("\n Empate!")

combate(itadori, nanami)
```
